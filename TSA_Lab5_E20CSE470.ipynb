```python
b,a = scipy.signal.butter(3,0.3,'lowpass')
filteredLowPass  = scipy.signal.filtfilt(b,a,df.meantemp)
```

```python
b,a = scipy.signal.butter(3,0.05,'highpass')
filteredHighPass  = scipy.signal.filtfilt(b,a,df.meantemp)
```

```python
b,a = scipy.signal.butter(3,[.01,0.5],'band')
filteredBandPass  = scipy.signal.lfilter(b,a,df.meantemp)
```

```python
fig = px.line(filteredBandPass, df.date, df.meantemp, title='BandPass')
fig.show()
```

```python
b,a = scipy.signal.butter(3,0.3,'lowpass')
filteredLowPass  = scipy.signal.filtfilt(b,a,df.humidity)
b,a = scipy.signal.butter(3,0.05,'highpass')
filteredHighPass  = scipy.signal.filtfilt(b,a,df.humidity)
b,a = scipy.signal.butter(3,[.01,0.5],'band')
filteredBandPass  = scipy.signal.lfilter(b,a,df.humidity)
fig = px.line(df, df.date, df.humidity, title='ECG Signal with noise')
fig.show()
fig = px.line(filteredLowPass, df.date, df.humidity, title='LowPass')
fig.show()
fig = px.line(filteredHighPass, df.date, df.humidity, title='HighPass')
fig.show()
fig = px.line(filteredBandPass, df.date, df.humidity, title='BandPass')
fig.show()
```

```python
time_series_avg = np.mean(df.humidity)
# create array composed of mean value and equal to length of time array
df['Simple Average Smoothing'] = np.full(shape=len(df.date), fill_value=time_series_avg, dtype='float')
fig = px.line(df, x=df.date, y=df.columns)
fig.show()
```

```python
df2 = df[['date', 'humidity']].copy()
df2['Moving_AVG'] = df['humidity'].rolling(window=3).mean()
fig = px.line(df2, df2.date, df2.columns, title='Equally Weighted Moving Average Smoothing')
fig.show()

MSE = np.square(np.subtract(df2.humidity, df2['Moving_AVG'])).mean()
print('Mean Square Error: ', MSE)
```

```python
weights = np.array([0.1, 0.2, 0.3])
df2['EWMA'] = df['humidity'].rolling(3).apply(lambda x: np.sum(weights*x))

fig = px.line(df2, df2.date, df2. columns, title="Exponential Weighted Moving Average Smoothing")
fig.show()

MSE = np.square(np.subtract(df2.humidity, df2['EWMA'])).mean()
print('Mean Square Error: ', MSE)
```
